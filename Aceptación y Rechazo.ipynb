Código para ejecutar el método, se van a generar valores para una variable "x" conociendo ya los valores de la variable "Y".

```python
from random import random

def geny():
  u = random()
  if u < 1/3:
    return 2
  elif u < 2/3:
    return 7
  else:
    return 9

def fx(x):
  if x == 2:
    return .2
  elif x == 7:
    return .5
  else:
    return .3

def fy(x):
  if x == 2:
    return 1/3
  elif x == 7:
    return 1/3
  else:
    return 1/3

c = 1.6
n = 10000
l = []
while len(l) < n:
  y = geny()
  u = random()
  if fx(y) / (c * fy(y)) <= u:
    l.append(y)

print(l)

print(l.count(2)/n,l.count(7)/n,l.count(9)/n)

```

$f_y (y)= \begin{cases} \dfrac{\ln(y)}{2\ln(2)-1}, & y \in[1,2]\\0, & \text{en otro caso} \end{cases}$

```python
from math import log
import matplotlib.pyplot as plt

x1=1
l=[]
while x1 <= 2:
  l.append(x1)
  x1=x1+.01
ly=[]
for x in l:
  ly.append(log(x)/(2*log(2)-1))
plt.plot(l,ly,color='yellow')
plt.show()
```

```python
def geny():
  return 1+random()
def fx(y):
  if x>1 and x<=2:
    return log(y)/(2*log(2)-1)
  else:
    return 0
def fy(x):
  if x>1 and x<=2:
    return 1
  else:
    return 0
c=1.6
n=1000000
l=[]
while len(l)<n:
  y=geny()
  u=random()
  while fx(y)/(c*fy(y))<=u:
    y=geny()
    u=random()
  l.append(y)
plt.hist(l,density=1, bins=50, color='orange', edgecolor='red')
plt.show()
```

```python
x1=1
b=[]
lt=[]
while x1<=2:
  b.append(x1)
  lt.append(fx(x1))
  x1=x1+1/50
plt.figure(figsize=(5,3))
plt.hist(l,density=1,bins=b,color='orange',edgecolor='red')
plt.plot(b,lt,color='yellow')
plt.show()
```
